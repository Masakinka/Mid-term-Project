Висунення гіпотез:
1. Гіпотеза: Чим довший останній контакт з клієнтом, тим вища ймовірність, що клієнт підпише депозит. (Тривалість контакту (duration) значно впливає на ймовірність підписання депозиту)
2. Гіпотеза: Клієнти з професійними або управлінськими посадами та вищою освітою частіше підписують депозит. (Тип роботи (job) та рівень освіти (education) впливають на ймовірність підписання депозиту)
3. Гіпотеза: Клієнти, які мали успішний досвід з попередніми кампаніями, мають вищу ймовірність підписання депозиту.(Результати попередніх маркетингових кампаній (poutcome) впливають на успіх поточної кампанії)
4. Гіпотеза: Покращення економічних умов сприяє підписанню депозиту.Економічні показники (emp.var.rate, euribor3m, nr.employed) впливають на рішення клієнтів)

Тому перед проджовждення проведу аналіз на мультиколінеарність допомагає виявити сильну кореляцію між незалежними змінними, що може призвести до проблем у моделюванні.

**Обчислення VIF (Variance Inflation Factor)**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Додавання константи до набору даних
X = add_constant(number_df)

# Обчислення VIF для кожної змінної
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

Загальні висновки:
- const: Дуже високий VIF для константи (528669.82) не є проблемою, оскільки це просто наслідок додавання константного стовпця.
- emp.var.rate: VIF дорівнює 33.07, що вказує на сильну мультиколінеарність.
euribor3m: VIF дорівнює 64.35, що також вказує на дуже високу мультиколінеарність.
- nr.employed: VIF дорівнює 31.68, що також вказує на сильну мультиколінеарність.
- Інші змінні мають VIF менше 10, що вказує на відсутність значної мультиколінеарності серед них.

1. Сильна мультиколінеарність. Змінні emp.var.rate, euribor3m та nr.employed мають дуже високі значення VIF, що вказує на сильну мультиколінеарність між ними.
2. Дії для вирішення проблеми мультиколінеарності:
- Видалення змінних: Оскільки ці змінні мають високу мультиколінеарність, можна розглянути можливість видалення однієї або кількох з них. Наприклад, можна видалити nr.employed або emp.var.rate.
- Зменшення розмірності, може допомогти зменшити мультиколінеарність.
Регуляризація: Використання моделей з регуляризацією (наприклад, Ridge або Lasso регресія) може допомогти впоратися з мультиколінеарністю.

**Для побудови моделей бінарної класифікації пропонується використовувати такі методи ML:**
1. Логістична регресія (Logistic Regression):
Переваги: Проста в реалізації, інтерпретована модель, добре працює для лінійно віддільних класів.
Недоліки: Може погано працювати з великою кількістю взаємозалежних змінних.
2.k-Nearest Neighbors (kNN):
Переваги: Інтуїтивно зрозуміла модель, добре працює для невеликих обсягів даних.
Недоліки: Може бути повільною для великих наборів даних, чутлива до масштабу змінних.
3. Дерево рішень (Decision Tree):
Переваги: Легка в інтерпретації, може обробляти як числові, так і категоріальні дані без особливого препроцесингу.
Недоліки: Схильна до перенавчання (overfitting), якщо не застосовувати обрізання (pruning).
4. Алгоритми бустингу:
Наприклад, XGBoost, LightGBM, CatBoost.
Переваги: Висока продуктивність, добре працює з великими наборами даних, може обробляти пропущені значення.
Недоліки: Складність налаштування гіперпараметрів.
5.Метрики оцінки якості моделей
Для оцінки якості моделей класифікації пропонується використовувати такі метрики:
- Accuracy (Точність):
Переваги: Легко зрозуміла.
Недоліки: Може бути недоречна для незбалансованих наборів даних.
- Precision (Точність) та Recall (Повнота):
Переваги: Добре підходять для оцінки незбалансованих наборів даних.
Недоліки: Точність може бути низькою, якщо мета - виявити всі позитивні випадки (висока повнота).
- F1-Score:
Переваги: Балансує між точністю та повнотою.
Недоліки: Може бути важко інтерпретувати, якщо метрики точності та повноти мають великий розрив.
- ROC AUC:
Переваги: Добре підходить для оцінки моделей незалежно від порогів.
Недоліки: Може бути важко інтерпретувати для бізнесу.

**Препроцесинг даних для подальшої їх передачі в модель**

Створемо кастомний трансформатор даних, щоб потім перевикористовувати для всіх моделей


```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
```

```python
# Трансформер для створення нових ознак на основі існуючих стовпців у DataFrame.
class FeatureCreationTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Створення додаткових ознак
        if 'balance' in X.columns and 'duration' in X.columns:
            X['balance_per_duration'] = X['balance'] / (X['duration'] + 1)  # +1 для уникнення ділення на нуль
        if 'campaign' in X.columns and 'previous' in X.columns:
            X['contacts_total'] = X['campaign'] + X['previous']
        print(f"FeatureCreationTransformer: {X.shape}")
        return X
```

```python
# Трансформер для обробки рідкісних категорій та виявлення і обробки аномальних значень у DataFrame.
class PreprocessingTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Групування рідкісних категорій
        X = X.copy()
        X['job'] = X['job'].replace(['housemaid', 'unknown', 'student', 'unemployed'], 'other')
        X['education'] = X['education'].replace(['illiterate', 'basic.4y', 'basic.6y', 'basic.9y'], 'basic')

        # Виявлення та обробка аномальних значень за допомогою IQR
        numeric_feats = X.select_dtypes(include="number").columns
        Q1 = X[numeric_feats].quantile(0.25)
        Q3 = X[numeric_feats].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        for feature in numeric_feats:
            initial_shape = X.shape[0]
            X[feature] = np.where(X[feature] < lower_bound[feature], lower_bound[feature], X[feature])
            X[feature] = np.where(X[feature] > upper_bound[feature], upper_bound[feature], X[feature])
            final_shape = X.shape[0]
            print(f"PreprocessingTransformer: {feature}, adjusted {initial_shape - final_shape} rows")
        print(f"PreprocessingTransformer: {X.shape}")
        return X
```

```python
# Розділення даних на тренувальний та тестовий набори
X = raw_df.drop('y', axis=1)
y = raw_df['y'].apply(lambda x: 1 if x == 'yes' else 0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Cпискb категоріальних та числових змінних
categorical_features = X_train.select_dtypes(include="object").columns.tolist()
numeric_features = X_train.select_dtypes(include="number").columns.tolist()
```

```python
# Пайплайн для препроцесингу числових змінних
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Пайплайн для препроцесингу категоріальних змінних
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(drop='first'))
])

# Об'єднання трансформерів у ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Створення пайплайнів для моделей з додаванням CustomTransformer
models = {
    'Logistic Regression': Pipeline(steps=[
        ('feature_creation', FeatureCreationTransformer()),
        ('preprocess', PreprocessingTransformer()),
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression(max_iter=1000))
    ]),
    'Decision Tree': Pipeline(steps=[
        ('feature_creation', FeatureCreationTransformer()),
        ('preprocess', PreprocessingTransformer()),
        ('preprocessor', preprocessor),
        ('classifier', DecisionTreeClassifier())
    ]),
    'k-Nearest Neighbors': Pipeline(steps=[
        ('feature_creation', FeatureCreationTransformer()),
        ('preprocess', PreprocessingTransformer()),
        ('preprocessor', preprocessor),
        ('classifier', KNeighborsClassifier())
    ]),
    'XGBoost': Pipeline(steps=[
        ('feature_creation', FeatureCreationTransformer()),
        ('preprocess', PreprocessingTransformer()),
        ('preprocessor', preprocessor),
        ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='logloss'))
    ])
}
```

```python
# Визначення імен функцій після виконання препроцесингу
def get_feature_names(column_transformer):
    output_features = []
    for name, pipe, features in column_transformer.transformers_:
        if isinstance(pipe, Pipeline):
            if 'onehot' in pipe.named_steps:
                # Get feature names for onehot encoding
                categories = pipe.named_steps['onehot'].categories_
                feature_names = []
                for cat, feature in zip(categories, features):
                    feature_names.extend([f"{feature}_{c}" for c in cat[1:]])  # Skip the first to match drop='first'
                output_features.extend(feature_names)
            else:
                output_features.extend(features)
        else:
            output_features.extend(features)
    return output_features
```

```python
# Встановлення імен функцій для FeatureCreationTransformer після виконання препроцесингу
for model in models.values():
    model.named_steps['preprocessor'].fit(X_train)
    feature_names = get_feature_names(model.named_steps['preprocessor'])
    feature_names.extend(['balance_per_duration', 'contacts_total'])  # Додаємо нові ознаки
    model.named_steps['feature_creation'].feature_names = feature_names

# Тренування та оцінка моделей
results = []
for model_name, model_pipeline in models.items():
    print(f"Training {model_name}...")
    model_pipeline.fit(X_train, y_train)
    y_train_pred = model_pipeline.predict(X_train)
    y_test_pred = model_pipeline.predict(X_test)

    train_report = classification_report(y_train, y_train_pred, output_dict=True)
    test_report = classification_report(y_test, y_test_pred, output_dict=True)

    results.append({
        'model': model_name,
        'train_precision': train_report['weighted avg']['precision'],
        'train_recall': train_report['weighted avg']['recall'],
        'train_f1': train_report['weighted avg']['f1-score'],
        'test_precision': test_report['weighted avg']['precision'],
        'test_recall': test_report['weighted avg']['recall'],
        'test_f1': test_report['weighted avg']['f1-score']
    })

# Виведення результатів у вигляді таблиці
results_df = pd.DataFrame(results)
print(results_df)
```

**Randomized Search з використанням Sklearn**

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from hyperopt import fmin, tpe, hp, Trials
from sklearn.metrics import f1_score
```

```python
# Параметри, використані для моделі XGBoost з початковими значеннями
initial_params = {
    'max_depth': 5,
    'learning_rate': 0.1,
    'n_estimators': 100,
    'colsample_bytree': 0.8,
    'subsample': 0.8
}

# Створення моделі з початковими параметрами
xgb_model = xgb.XGBClassifier(
    max_depth=initial_params['max_depth'],
    learning_rate=initial_params['learning_rate'],
    n_estimators=initial_params['n_estimators'],
    colsample_bytree=initial_params['colsample_bytree'],
    subsample=initial_params['subsample'],
    use_label_encoder=False,
    eval_metric='logloss'
)

# Параметри для Randomized Search
param_dist = {
    'max_depth': range(3, 10),
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],
    'n_estimators': range(50, 500, 50),
    'colsample_bytree': [0.3, 0.5, 0.7, 0.8, 0.9, 1.0],
    'subsample': [0.3, 0.5, 0.7, 0.8, 0.9, 1.0]
}

# Randomized Search
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,
    scoring='f1_weighted',
    cv=3,
    verbose=1,
    random_state=42
)

# Виконання Randomized Search
random_search.fit(preprocessor.fit_transform(X_train), y_train)
print("Best parameters found by Randomized Search:")
print(random_search.best_params_)

# Оцінка моделі з Randomized Search
best_random_search_model = random_search.best_estimator_
y_pred_random_search = best_random_search_model.predict(preprocessor.transform(X_test))
f1_random_search = f1_score(y_test, y_pred_random_search, average='weighted')
print(f"F1 Score for Randomized Search: {f1_random_search}")
```

```python
# Простір пошуку для Hyperopt
space = {
    'max_depth': hp.choice('max_depth', range(3, 10)),
    'learning_rate': hp.uniform('learning_rate', 0.01, 0.3),
    'n_estimators': hp.choice('n_estimators', range(50, 500, 50)),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.3, 1.0),
    'subsample': hp.uniform('subsample', 0.3, 1.0)
}

# Цільова функція для Hyperopt
def objective(params):
    model = xgb.XGBClassifier(
        max_depth=params['max_depth'],
        learning_rate=params['learning_rate'],
        n_estimators=params['n_estimators'],
        colsample_bytree=params['colsample_bytree'],
        subsample=params['subsample'],
        use_label_encoder=False,
        eval_metric='logloss'
    )
    model.fit(preprocessor.transform(X_train), y_train)
    y_pred = model.predict(preprocessor.transform(X_test))
    f1 = f1_score(y_test, y_pred, average='weighted')
    return -f1

# Запуск Hyperopt
trials = Trials()
best_params_hyperopt = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=50, trials=trials, rstate=np.random.default_rng(42))
print("Best parameters found by Hyperopt:")
print(best_params_hyperopt)

# Створення та оцінка моделі з Hyperopt
best_model_hyperopt = xgb.XGBClassifier(
    max_depth=best_params_hyperopt['max_depth'],
    learning_rate=best_params_hyperopt['learning_rate'],
    n_estimators=best_params_hyperopt['n_estimators'],
    colsample_bytree=best_params_hyperopt['colsample_bytree'],
    subsample=best_params_hyperopt['subsample'],
    use_label_encoder=False,
    eval_metric='logloss'
)
best_model_hyperopt.fit(preprocessor.transform(X_train), y_train)
y_pred_hyperopt = best_model_hyperopt.predict(preprocessor.transform(X_test))
f1_hyperopt = f1_score(y_test, y_pred_hyperopt, average='weighted')
print(f"F1 Score for Hyperopt: {f1_hyperopt}")
```
